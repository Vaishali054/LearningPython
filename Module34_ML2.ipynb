```python
#Nominal /OHE - binary vector for each category

import pandas as pd

from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'status':['single', 'married', 'single', 'separated']})
df
```

```python
encoder=OneHotEncoder()
```

```python
encoder
```

```python
encoded=encoder.fit_transform(df[['status']]).toarray()
```

```python
 encoder_df=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
 encoder_df
```

```python
encoder.transform([['single']]).toarray()
```

```python
df
```

```python
pd.concat([df,encoder_df], axis=1)
```

```python
#Ordinal Encoding

from sklearn.preprocessing import OrdinalEncoder

df=pd.DataFrame({'qualification': ["HS","GR","PHD","HS", "HS"]})
```

```python
df
```

```python
encoder=OrdinalEncoder(categories=[['HS','GR','PHD']])
```

```python
encoder.fit_transform(df[['qualification']])
```

```python
#new data
encoder.transform([['PHD']])
```

```python
#Target guided ordinal encoding

df=pd.DataFrame({'time': ['lunch', 'breakfast', 'dinner', 'lunch'], 'total_bill' :[100, 200, 250, 400]})
```

```python
mean_value=df.groupby('time')['total_bill'].mean().to_dict()
```

```python
df['time_encoded']=df['time'].map(mean_value)
```
